# Single inflation
## Background Dynamics
We consider a model consisting of one scalar fields, say, $\phi$, whose dynamics is governed by the following action:
$$
S=\int d^4x\sqrt{-g}\left[\frac{M_\mathrm{P}^2}{2}R-\frac{1}{2}(\partial\phi)^2-V(\phi)\right]\,,
$$
where $\phi$ is a canonical scalar field and $M_\mathrm{P}$ is reduced  Plank mass. We consider the following potential:
$$
V=\frac{1}{2}m^2\phi^2\,.
$$
We work with the spactially flat Friedmann-Robertson-Walker (FRW) universe by the line-element
$$
ds^2=-dt^2+a^2(t)dx^2\,,
$$
where $a(t)$ is the scale factor and $t$ is the cosmic time. In this background, the equation of motion (EOM) is given by (Euler-Lagrange equation):

$$\ddot \phi=-3H\dot\phi-V_{,\phi}=-3H\dot\phi-m^2\phi\,,$$
where $H$ is the Hubble parameter and the overdots represent differentiation with respect to the cosmic time. The dynamics of the scale factor is described by the following Friedmann equations:

$$H^2=\frac{1}{3M_\mathrm{P}^2}\left[\frac{1}{2}\dot\phi^2+V\right]\,,$$

$$\dot H=-\frac{\dot\phi^2}{2M_\mathrm{P}^2}\,.$$

At this point, we can solve equations and obtain trajectory. But in order to allow e-folds and scale factor to appear naturally, we need to redefine these dimensionless variables, which will facilitate our subsequent processing of the power spectrum.


$$T=t\cdot M_\mathrm{P} \qquad x=\frac{\phi}{M_\text{P}}$$

$$y=\frac{\dot\phi}{M_\mathrm{P}^2}\qquad  z=\frac{H}{M_\mathrm{P}}\qquad a=a\,,$$

EOM and sacle factor Friedmann Equations become:

$$\frac{dx}{dT}=y$$

$$\frac{dy}{dT}=-3zy-\hat V_\phi$$

$$\frac{dz}{dT}=-\frac{1}{2}y^2 $$

$$\frac{da}{dT}=az$$

where we define $\hat V=\frac{m^2\phi^2}{2M_\rm P^4}$ and $\hat V_\phi=\frac{V_\phi}{M_{\rm P}^3}=\frac{m^2\phi}{M_\rm P^3}=\frac{m^2x}{M_\rm P^2}$, see code for details.

## Another way (e-folds) to express the equation of motion

By taking the derivative of the above equation with respect to the number of e-folds $N=\ln a$, we obtain

$$\dot\phi=\frac{d}{dt}\phi=\frac{d}{dN}\frac{dN}{dt}\phi=H\phi'\,,$$

$$\ddot\phi=\frac{d}{dt}(H\phi')=HH'\phi'+H^2\phi''\,.$$

where, as usual, the primes represent differentiation with respect to the e-folds. Then the equation of motion, Friedmann equations and scale factor become:

$$\phi''=-(3+\frac{H'}{H})\phi'-\frac{V_\phi}{H^2}=-(3-\frac{1}{2}\phi'^2)\phi'-\frac{m^2\phi}{H^2}\,,$$

$$H'=\frac{\dot H}{H}=-\frac{1}{2}H\phi'^2\,,$$

$$a'=\frac{d a}{d lna}=a\,.$$
We also define the dimensionless parameters
$$N=N \qquad x=\frac{\phi}{M_\text{P}}$$

$$y=\frac{\phi\prime}{M_\text{P}}\qquad  z=\frac{H}{M_\text{P}}\qquad a=a\,.$$
EOM, sacle factor and  Friedmann Equations become
$$\frac{dx}{dN}=y\,,$$

$$\frac{dy}{dN}=-(3-\frac{1}{2}y^2 )y - \frac{\hat V_\phi}{z^2}\,,$$

$$\frac{dz}{dN}=-\frac{1}{2}zy^2\,,$$

$$\frac{da}{dN}=a\,.$$
where we define dimensionless potential $\hat V=\frac{V}{M_\rm P^4}$ and $\hat V_\phi=\frac{V_\phi}{M_\rm P^3}=\frac{m^2\phi}{M_\rm P^3}=\frac{m^2x}{M_\rm P^2}$
## Slow roll parameters

$$\epsilon_H=-\frac{\dot H}{H^2}=\frac{1}{2}\phi\prime^2=1-\frac{a\ddot a }{\dot a^2}\qquad \eta_H=\frac{\dot \epsilon_H}{H\epsilon_H}=\frac{\epsilon_H'}{\epsilon_H}=\frac{2\phi^{\prime\prime}}{\phi'}$$

Accelerated expansion occurs if $\epsilon_H<1$. The de Sitter limit, $p_\phi\sim−\rho_\phi$, corresponds to $\epsilon_H\sim 0$. In this case, the potential energy dominates over the kinetic energy.
$\eta_H$ starts small and then oscillates with the subsequent oscillation of $\epsilon_H$.where $|\eta_H|\ll 1$ ensures that the fractional change of $\epsilon_H$ per e-fold is small. The spacetime is approximately de Sitter $a\sim e^{Ht}$, when $\epsilon_H, |\eta_H|\ll 1$. 
$$H^2\approx \frac{1}{3}V(\phi)\sim \rm const \qquad \dot\phi\approx\frac{V_\phi}{3H}$$
Inflation ends when the slow-roll conditions are violated
$$\epsilon_H(\phi_{\rm end})=1\qquad\eta_H(\phi_{\rm end})\approx1 $$
The number of e-folds before inflation ends is
$$N(\phi)=\ln\frac{a_\rm {end}}{a}=\int^{t_{\rm end}}_{t}Hdt=\int^{\phi_{\rm end}}_{\phi}\frac{H}{\dot\phi}d\phi\approx\int_{\phi_{\rm end}}^{\phi}\frac{V}{V_\phi}d\phi=\int_{\phi_{\rm end}}^{\phi}\frac{d\phi}{\sqrt{2\epsilon_V}}\,.$$
Some constraints suggest that inflation should continue for 60 e-folds.  The slow-roll conditions,$\epsilon_H,|\eta_H|\ll 1$, may also be expressed as conditions on the shape of the inflationary potential ($\dot\phi\sim\frac{V_\phi}{3H},H^2\sim\frac{V}{3M_{\rm P}}$)
$$\epsilon_V(\phi)=\frac{M_\rm P ^2}{2}\left(\frac{V_\phi}{V}\right)^2\qquad \eta_V(\phi)=M_{\rm P}^2\frac{V_{\phi\phi}}{V}$$
## Selection of initial values
Regarding the choice of initial values, we use the Planck unit system, $M_{\rm P}=1$. 
Field $\phi_i$ should be chosen to be around 15, so that inflation can last for 60 efolds.
$$N(\phi)=\int_{\phi_{\rm end}}^{\phi_i}\frac{V}{V_\phi}d\phi=\int_{0}^{\phi_i}\frac{1}{2}\phi d\phi=\frac{1}{4}\phi^2_i\sim 60\,.$$
To satisfy the normalization of scalar fluctuations, $\Delta^2_s=10^{-9}$, we need to fix the inflaton mass to $m\sim 10^{-6}M_{\rm P}$
$$\Delta^2_s(k)=\frac{1}{24\pi^2}\frac{V}{M_\rm{P}^4}\frac{1}{\epsilon_V}=\frac{m^2\phi^4}{96\pi^2 M_{\rm P}^6}\sim 10^{-9}$$
considering Slow roll approximation, we have
$$|\ddot \phi|\ll |3H\dot\phi|\,,|V_\phi|$$
The EOM can be written as:
$$0+3H\dot\phi+V_\phi=0\qquad \dot\phi=-\frac{V_\phi}{3H}=\frac{m^2\phi}{3H}\qquad y_i=-\frac{m^2x}{3z_iM_{\rm P}^2}$$
Friedman equation and the initial value of $z$:
$$H^2=\frac{1}{3M_{\rm P}}\left(\frac{1}{2}\dot\phi^2+V\right) \qquad z_i=\sqrt{\frac{1}{6}\left(y_i^2+\frac{m^2x_i^2}{M_{\rm P}^2}\right)}$$
The initial value can be summarized as：
$$M_\rm P=1\qquad m=10^{-6}M_\rm P \qquad x_i=15\qquad a_i=1$$
$$
\begin{cases} \frac{\dot\phi}{M_{\rm P}}=y_i=-\frac{m^2x}{3z_iM_{\rm P}^2} \\
\frac{H}{M_{\rm P}}=z_i=\sqrt{\frac{1}{6}\left(y_i^2+\frac{m^2x_i^2}{M_{\rm P}^2}\right)}\\
\end{cases} 
$$
Note that here $y_i$ is the derivative of the universe time. If it is the derivative of N, we need $y_i=\frac{y_i}{z_i}$. Considering the real part, we have:

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
from sympy import symbols, Eq, solve
m_p=1
m = 1e-6*m_p
def Vp(x):# this is dimensionless \hat V=V/(M^4_P)
    return 0.5*m**2*x**2
def V_phi(x):# this is dimensionless \hat V_phi=V_phi/(M^3_P)
    return m**2*x
# defination above should be changed for different models
y, z = symbols('y z')
x = 15
a=1
eq1 = Eq(y, -V_phi(x) / (3 * z))#yi equation
eq2 = Eq(z, ((y**2 + 2*Vp(x)) / 6)**0.5)#zi equation
solutions = solve((eq1, eq2), (y, z))
print(solutions)
xi=x
ai=a
yi,zi=solutions[0]
yi=yi/zi
print(f'The initial value of x: {xi}\nThe initial value of y: {yi} \nThe initial value of z: {zi}\nThe initial value of a: {ai}')
```

```python
def sys(t, var):#here, t actully is N, to avoid naming trouble later, we will briefly use t here.
    x, y, z, a = var
    dxdN = y
    dydN = -(3 - 0.5 * y**2) * y - V_phi(x) / (z**2)
    dzdN = -0.5 * z * y**2
    dadN = a
    return [dxdN, dydN, dzdN, dadN]

N = np.linspace(0, 65, 10000)  # Time points where solution is needed

# invoking the ODE solver
sol = solve_ivp(sys, [N[0], N[-1]], [xi, yi, zi, ai], t_eval=N, rtol=1e-8, atol=1e-10, method='RK45')
t = sol.t
x, y, z, a = sol.y
phi, phi_N, H = x*m_p, y*m_p, z*m_p

plt.plot(phi, phi_N, 'r')
plt.xlabel(r'$\phi$')
plt.ylabel(r'$\phi\prime $')
plt.show()

plt.plot(t, phi, label=r'$\phi$')
plt.plot(t, phi_N, label=r'$\phi\prime$')
plt.legend()
plt.xlabel('N')
plt.show()
```

```python
eps=0.5*phi_N**2

plt.plot(N, eps, label=r'$\epsilon_H$')
plt.axhline(1, color='black')
plt.legend()
plt.xlabel('N')
plt.show()

index_of_inflation_end = np.argmin(np.abs(eps - 1)) 
N_value=N[index_of_inflation_end]
print(f'The end of inflation: N= {N_value}')


phi_NN=-(3 - 0.5 * y**2) * y - m**2 * x / (z**2*m_p**2)
#To avoid singularities in the function, we exclude points where phi_N = 0, and to prevent strong oscillations following the end of inflation, we only present the results for the three e-folds after the end of inflation.
# eta=-3-m**2*phi[1:]/(H[1:]**2*phi_N[1:])
eta=2*phi_NN/phi_N
plt.plot(N, eta, label=r'$\eta_H $')
plt.legend()
plt.xlabel('N')
plt.xlim(1, N_value)
plt.ylim(-1,1)
plt.show()
```

## Analyze the behavior of $\phi$ and $\dot\phi$ 

Let's think analytically. Combine EOM and Friedman equations:

$$\frac{d\dot\phi}{d\phi}=-\frac{\sqrt{\frac{3}{2}}\left(\dot\phi^2+m^2\phi^2\right)^{1/2}\dot\phi+m^2\phi}{\dot\phi}\,,$$

say, Eq.1

*1)  $\dot\phi\gg m^2\phi^2$ 

It describes the situation when the potenial energy is small compared to the kinetic energy. Neglecting $m\phi$ compared to $\dot\phi$ in Eq.1, we obtain
  
$$\frac{d\dot\phi}{d\phi}=-\sqrt{\frac{3}{2}}\dot\phi\,.$$

The solution of this equation is 
        
$$\dot\phi=\dot\phi_0 e^{-\sqrt{\frac{3}{2}}\phi}\,,$$

$$\phi=\text{const}-\sqrt{\frac{2}{3}}\ln t$$
where $\dot\phi_0<0$ is a constant of integration. 

$\dot\phi$ decays exponentially, and regardless of the initial value, it will rapidly meet the attractor. During this process, $\phi$ does not undergo significant changes.

$$\phi\sim lnt\qquad H^2\sim t^{-2}$$

However, in the context of slow-roll inflation, we generally do not consider scenarios with significant kinetic energy at begining.
    
*2)   $\dot\phi\ll m^2\phi^2$, $|\phi|\gg 1$

The afterwards teh solution decribes a stage of accelerated expansion. To determine the attractor solution we assume that $d\dot\phi/d\phi\approx 0$.

Then the Eq.1 becomes:
$$
0=-\frac{\sqrt{\frac{3}{2}}\left(m^2\phi^2\right)^{1/2}\dot\phi+m^2\phi}{\dot\phi}=\sqrt{\frac{3}{2}}+\frac{m}{\dot\phi}\,,
$$


$$\dot\phi\sim-\sqrt{\frac{2}{3}}m\sim const \qquad \phi\sim \sqrt{\frac{2}{3}}m\cdot(t_f-t)\,.$$

During this stage, $\dot\phi$ remains nearly constant while $phi$ continues to decay. Substituting the expression for $phi$ into the Friedmann equation (Neglecting kinetic term)

$$H^2=\left(\frac{\dot a}{a}\right)^2=\frac{1}{3}\left(\frac{1}{2}m^2\phi^2\right)=\frac{1}{3}\left(\frac{1}{2}m^2\frac{2m^2}{3}(t_f-t)^2\right)=\frac{1}{9}m^4(t_f-t)^2\,,$$

$$\Delta N=\frac{1}{4}\phi_i^2\,,$$

where $\phi_i$ is the initial value of field $\phi$. We can estimate the duration of the inflationary phase; for example, with $x = 15$ as set above, this corresponds to more than 50 e-folds. More details see Ref.[1]

*3)Graceful exit (more details show in Boltzmann equation.ipynb)

Friedman equation can be written as

$$6H^2\sim \dot\phi^2+m^2\phi^2\,.$$
We define new independent variables
$$\dot\phi\sim H sin\theta \qquad m\phi\sim Hcos\theta\,,$$
where $H$ is Hubble parameter and $\theta$ is angular variable.

It is convenient to replace Eq.1 by a system of two first order differential equations for $H$ and $\theta$.
$$\dot H\sim H^2 sin^2\theta\qquad \dot \theta=-m-\frac{3}{2}Hsin2\theta\,,$$
because $\dot H$ is always negative. H decays continuously, so the second of $\dot\theta$ term can be ignored. Then we obtain

$$\theta\sim -mt$$
where the intergal constant can ve set to zero. Taking $\theta$ into equation $H$

$$\dot H\sim -H^2 sin^2(-mt)$$
$$-\frac{dH}{H^2}\sim sin(mt)dt$$
$$\frac{1}{H}\sim \frac{1}{m}(\frac{mt}{2}-\frac{1}{4}sin2mt)$$
$$H\sim \frac{1}{t}\left(1-\frac{sin(2mt)}{2mt}\right)^{-1}\,.$$
Then, we can obtain equation of $\phi$
$$\phi\sim \frac{Hcos\theta}{m}\sim\frac{cosmt}{mt}\left(1+\frac{sin2mt}{2mt}\right)\,,$$

more detail we discuss in Boltzman equaiton.ipynb.
